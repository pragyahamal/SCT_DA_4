```python
import pandas as pd
import zipfile

with zipfile.ZipFile('/content/archive (3).zip', 'r') as zip_ref:
    zip_ref.extractall('/content/data')

df = pd.read_csv('/content/data/marketing_campaign.csv', sep=';')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Income'], bins=30, kde=True)

plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')

plt.show()
```

```python
products = ['MntWines', 'MntFruits', 'MntMeatProducts',
            'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

df[products].sum().plot(kind='bar', figsize=(8,5))

plt.title('Total Spending on Products')
plt.ylabel('Amount Spent')

plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_df.corr(), cmap='coolwarm')

plt.title('Correlation Heatmap')

plt.show()
```

```python
sns.countplot(x='Response', data=df)

plt.title('Campaign Response Count')

plt.show()
```
